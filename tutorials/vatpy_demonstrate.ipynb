```python
import numpy as np
import matplotlib.pyplot as plt

import vatpy as vp
```

```python
file = '/srv/astro/projects/zoom/jpeterss/accretion/dwarf/highres/h1b4_SNRT/OUTPUT/snap_194.hdf5'
tp = vp.TerminalPlot(file=file, unitlength='kpc')
```

```python
tp.density(box=(48, 52), bins=200)
```

```python
im_data = vp.get_image_data(file=file, quantity=['mass'], box=(48, 52))
im_data_zoom = vp.get_image_data(file=file, quantity=['mass'], box=(-0.1, 0.1), blackholefocus=True)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.imshow(np.log10(im_data['mass'].T), extent=im_data['extent'])

ax_zoom = ax.inset_axes([0.5, 0.5, 0.47, 0.47], xlim=(49.9, 50.1), ylim=(49.9, 50.1), xticklabels=[], yticklabels=[])
ax_zoom.imshow(np.log10(im_data_zoom['mass'].T), extent=im_data_zoom['extent'], origin="lower")

ax.indicate_inset_zoom(ax_zoom, edgecolor="black")

plt.show()
```
